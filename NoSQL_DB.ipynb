## NoSQL - Not Only SQL

## Причины возникновения NoSQL 
Ключевым фактором, заставившим мировое IT-сообщество задуматься над новыми стратегиями хранения и доступа к информации, стал планомерный рост объемов данных в сети Интернет. В связи с этим был создан термин Big Data, включающий в себя некую стратегию, позволяющую эффективно работать с огромными постоянно растущими массивами данных. И на фоне этой концепции четко вырисовывалась необходимость в модели базы данных, которая будет больше нацелена на:
<ol>
    <li>Скорость доступа</li>
    <li>Масштабируемость</li>
</ol>

## Особенности NoSQL
<ol>
    <li>Объект данных - более сложная структура, чем просто строка в таблице</li>
    <li>Без строгого определения схемы</li>
    <li>Без операции соединения <b>JOIN</b></li>
    <li>Масштабируется</li>
    <li>Нет <b>SQL</b>, но бывает что-то похожее</li>
</ol>

## CAP - теорема (распределенные системы)
<b>Теорема: </b> - для распределенных систем можно выбрать только <b>два свойства</b>
<ol>
    <li><b>Consistency</b> - информация на разных узлах согласована (пример: при проведении транзакция информация по ней появляется на всех серверах)</li>
    <li><b>Availibility</b> - система отвечает на запросы (при сбое БД может быть недоступна)</li>
    <li><b>Partition tolerance</b> -  связи между узлами могут обрываться</li>
</ol>
В случае с <code>NoSQL</code> выбирают <b>consistency</b> и <b>partition tolerance</b>
ктура

## BASE - архитектура (распределенные системы)
<ol>
    <li><b>Basic Available</b> - сбой узла приводит к отказу только для части пользователей</li>
    <li><b>Soft-State</b> - система может находиться в неустойчивом состоянии (пример такого состояния - незавершенная транзакция)</li>
    <li><b>Eventual Consistency</b> -  когда-нибудь согласуется</li>
</ol>


## Виды NoSQL СУБД
### 1. Ключ - значение
Используется для:
<ol>
    <li>Работы с данными в реальном времени (зачастую на SSD или оперативной памяти) </li>
    <li>Кэширование результатов других операций</li>
    <li>Файловые системы на основе "ключ-значение"</li>
</ol>
<b>Пример:</b> Redis, Memcache, Berkeley DB

### 2. Документоориентированные
Используется для:
<ol>
    <li>Хранение документов и поиска по ним</li>
    <li>Информационные сервисы</li>
    <li>Данные с нечеткой схемой</li>
</ol>
<b>Пример:</b> MongoDB, CouchDB

### 3. Семейство столбцов
Используется для:
<ol>
    <li>Хранение данных, собранных с веб-страниц</li>
    <li>Хранение и поиск по большому объему данных</li>
</ol>
<b>Пример:</b> Cassandra, HBase

## Redis

### Введение
<b>Redis</b> хранит всю БД в оперативной памяти, при этом периодически делает слепки и хранит на жестком диске, чтобы в случае сбоя не потерять все данные.
<ul>
    <li>100k операций в секунду для одного узла</li>
    <li>Больше сотни команд</li>
    <li>Все команды работы с данными содержат ключ</li>
    <li>Для каждого типа данных свой набор CRUD-команд, различаются префиксами, например LPOP, SPOP - достать элемент с конца List и Set соответственно</li>
</ul>

### Типы данных в Redis
<b>Ключ</b> - уникальная строка для одного набора данных
<b>Значение:</b>
<ul>
    <li><b>Строки</b> - скалярные значения, для разных подтипов есть разные операции</li>
    <li><b>Хэши</b> - для хэш-таблиц внутри значений</li>
    <li><b>Списки</b> - как список в Python</li>
    <li><b>Множества</b> - как множество в Python</li>
    <li><b>Упорядоченные множества</b> 
</ul>

### Репликация в Redis
<ul>
    <li>Обновление поступают не только от мастера, но и от уже обновленных <b>slave'ов</b>, децентрализуя операцию</li>
    <li>Изменения всегда просты (с одним ключом), поэтому передаются значения, а не операторы</li>
</ul>
<b>Подписки (PubSub)</b>: не нужно каждый раз спрашивать есть ли новые данные, сервер сам пройдется по активным подписчикам и отправит изменения, когда они появятся.
