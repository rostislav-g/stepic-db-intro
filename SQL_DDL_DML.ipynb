## SQL / DDL
<b>DDL</b> - это язык (подмножество языка SQL), который служит для создания и модификации структура БД (создание / изменение / удаление таблиц и связей).<br><br>

### Создание схемы
В MySQL: СХЕМА = БД <br>

<b>Синтаксис:</b><br>
<code><i><schema definition\></i> ::=
    CREATE SCHEMA <i><schema name clause\></i>
    [ <i><schema character set specification\></i> ]
    [ <i><schema element\></i>... ]</code><br>        

<b>Пример:</b><br>
<code>CREATE SCHEMA IF NOT EXISTS 'store' DEFAULT CHARACTER SET utf8</code></li><br><br>

### Создание домена
<b>Домен</b> - множество допустимых значений (в рамках MySQL не предусмотрено)<br>
<b>Синтаксис:</b><br>
<code><i><domain definition\></i> ::= CREATE DOMAIN <i><domain name\></i> AS data_type
    [ <i><default definition\></i> ]
    [ <i><domain_constraint_definition_list\></i> ]    
<b>default_definiton</b> ::= DEFAULT ( literal | niladic_fucntion | NULL)
<b>domain_constraint_definition_list</b> ::= [CONSTRAINT constraint_name]
        CHECK (condition_expression)</code><br><br>
<b>Пример (PostgreSQL):</b><br><br>
<code>CREATE DOMAIN SALE_STATUS AS VARCHAR(45)
    DEFAULT 'new'
    CHECK (VALUE IN ('new', 'process', 'assembly', 'ready', 'issued')
    CONSTRAINT SALE_STATUS_NOT_NULL CHECK (VALUE IS NOT NULL);</code>
